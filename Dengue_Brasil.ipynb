Calcular a quantidade de pessoas que contrairam dengue por municípo e por ano

```python
#Criar uma nova coluna de ano
df_dengue ['ano'] = df_dengue ['data_infeccoes'].dt.year
```

```python
df_dengue.head()
```

```python
infeccoes_municipios = df_dengue.groupby(['ano','municipio','uf'])['qtd_infeccoes'].sum().reset_index()
```

```python
infeccoes_municipios.head()
```

```python
infeccoes_municipios.info()
```

Agrupar por estado e ano

```python
infeccoes_estado = df_dengue.groupby(['ano','uf',])['qtd_infeccoes'].sum().reset_index()
```

```python
infeccoes_estado.head(15)
```

Estatística Descritiva

```python
infeccoes_estado.describe()
```

```python
infeccoes_estado.shape
```

Interpretação da estatística descritiva utilizando o gráfico de Boxplot

```python
plt.figure(figsize = (10,6))
sns.boxplot(data = infeccoes_estado, y = 'qtd_infeccoes')
```

```python
#Criando boxplot comparando a quantidade de infecções por ano
plt.figure(figsize = (10,6))
sns.boxplot(data = infeccoes_estado, x = 'ano', y = 'qtd_infeccoes', palette = 'viridis')
 ## palette = pastel, viridis

#Personalizando o gráfico
plt.title('Quantidade  de infecções por ano', fontsize = 14)
plt.xlabel('Ano', fontsize = 12)
plt.ylabel('Quantidade de infecções', fontsize = 12)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)

plt.show()
```

Aprofundando o olhar para o ano de 2023

```python
#Filtrar os dados de infecções para o ano de 2023
infeccoes_estado_2023 = infeccoes_estado[infeccoes_estado['ano'] == 2023]
infeccoes_estado_2023
```

```python
infeccoes_estado_2023.describe()
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(data = infeccoes_estado_2023, y = 'qtd_infeccoes', palette = 'viridis')
plt.title('Boxplot de infecções por Estado / 2023', fontsize = 14)
plt.ylabel('Quantidade de infecções', fontsize = 12)
plt.grid(True)
plt.show()
```

```python
#Lib de visualização Plotly
import plotly.express as px
```

```python
fig = px.box(infeccoes_estado_2023, x = 'uf', y = 'qtd_infeccoes', color = 'uf', title = 'Boxplot de infecções por Estado / 2023')
fig.show()
```

```python
fig = px.box(infeccoes_estado_2023, y = 'qtd_infeccoes', title = 'Boxplot de infecções por Estado / 2023')
fig.show()
```

Gráfico de barras para visualização de comparação entre os Estados

```python
#Criar gráfico de barras comparando a quantidade de infecções por estado em 2023
plt.figure(figsize = (12,6))
sns.barplot(data = infeccoes_estado_2023, x = 'uf',  y = 'qtd_infeccoes', palette = 'viridis')

#Personalizar o gráfico
plt.title('Quantidade de infecções por Estado em 2023', fontsize = 14)
plt.xlabel('Estado', fontsize = 12)
plt.ylabel('Quantidade de infecções', fontsize = 12)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()
```

Ordenando os dados do maior para o menor de infecções

```python
infeccoes_2023 = infeccoes_estado_2023.sort_values(by = 'qtd_infeccoes', ascending=False )
```

```python
infeccoes_2023
```

```python
#Criar gráfico de barras comparando de forma ordenada
plt.figure(figsize = (12,6))
sns.barplot(data = infeccoes_2023, x = 'uf',  y = 'qtd_infeccoes', palette = 'viridis')

#Personalizar o gráfico
plt.title('Quantidade de infecções por Estado em 2023', fontsize = 14)
plt.xlabel('Estado', fontsize = 12)
plt.ylabel('Quantidade de infecções', fontsize = 12)
plt.xticks(rotation = 45)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()
```

Gráfico de tendências

```python
#Criar o gráfico de tendência
plt.figure(figsize = (12,6))
sns.regplot(data = infeccoes_estado, x = 'ano',  y = 'qtd_infeccoes',scatter_kws = {"s":15}, line_kws = {"color":"red"})

#Personalizar o gráfico
plt.title('Tendência de infecções ao longo dos anos', fontsize = 14)
plt.xlabel('Ano', fontsize = 12)
plt.ylabel('Quantidade de infecções', fontsize = 12)
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```
